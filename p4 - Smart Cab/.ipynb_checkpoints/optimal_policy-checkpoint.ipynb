```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('agent.csv')
```

```python
pd.DataFrame({'count' : df.groupby( ["Iterations"] ,as_index=False).size()})

df
```

```python
df[df['Iterations']==44]
```

```python
actions = pd.DataFrame({'Total_Actions':df.groupby(["Iterations"])[['Iterations']].count().to_dict()['Iterations']})
actions
```

```python
reward = pd.DataFrame(df.groupby(["Iterations"])[['Reward']].sum().to_dict())
reward
```

```python
penalty=pd.DataFrame(df.groupby(["Iterations"])[['Penalty']].sum().to_dict())
penalty
```

```python
result = pd.concat([reward, penalty,actions], axis=1, join_axes=[reward.index])
result['Iterations']=result.index
result
```

```python
final_df = pd.merge(result, df, how='inner', on=['Iterations']).sort_values(by=['Reward_x', 'Penalty_x','Total_Actions'], ascending=[0,1,1])
final_df.rename(columns={'Reward_x': 'Total_Rewards', 'Penaly_x': 'Total_Penalty','Reward_y':'Reward','Penalty_y':'Penalty'}, inplace=True)
```

```python
final_df[final_df['Iterations']==40]#.to_csv('optimal_policy.csv')
```

```python
d = final_df[final_df['Action']!=final_df['Waypoint']].sort_values(by=['Iterations','Trails'], ascending=[1,1])
```
